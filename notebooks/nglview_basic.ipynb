# Nglview basic usage

## Import the Nglview library

```python
import nglview as nv
from ase.build import molecule
```

## Construct the fullerene molecule from ASE

```python
c60 = molecule("C60")
c60.set_cell([[20, 0, 0],[0, 20, 0],[0, 0, 20]])
c60.center()
c60.pbc=True
```

## Show the molecule

```python
view = nv.show_ase(c60)
view.add_unitcell()
view
```

## Render image

```python
view.render_image()
view._display_image()
```

## Add unitcell

```python
view.add_unitcell()
view
```

## Change atom color and spacefill 

Here, I change all the carbon atom with blue color. 

```python
view.add_representation('spacefill', selection='#C', color='blue', radius='1.2')
view
```

## Label the atoms 

Here, I label the atoms with atom numbers and atom elements. 

```python
view.add_label(color='red', scale=1.3, labelType='text', labelText = [c60[i].symbol + str(i) for i in range(60)], zOffset=2.0, attachment='middle_center')
view
```
